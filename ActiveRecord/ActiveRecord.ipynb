ActiveRecord es una forma de interactuar con bases de datos relacionales, como MySQL o SQLite, usando lenguaje Ruby en lugar de SQL. Entonces cada tabla de una base de datos se convierte en una clase Ruby, y las filas de esa tabla se convierten en instancias de esa clase.

## Contenido del proyeto

- En la carpeta lib es donde reciden los archivos de codigo
- En la carpeta spec contiene las pruebas de RSpec
- customers.sqlite3 es la base de datos que contiene los clientes falsos
- customers.csv es una version de la lista de clientes
- Guardfile es para ejecutar automaticamente las pruebas
- Gemfile y Gemfile.lock son las dependencias (archivos de configuracion)

## Solucion de las pruebas

Estudiemos primero el archivo `activerecord_practica.rb`

```ruby
require 'sqlite3'
require 'active_record'
  require 'byebug'


ActiveRecord::Base.establish_connection(:adapter => 'sqlite3', :database => 'customers.sqlite3')
# Mostrar consultas en la consola.
# Comenta esta linea para desactivar la visualizacin de consultas SQL sin formato.
ActiveRecord::Base.logger = Logger.new(STDOUT)

class Customer < ActiveRecord::Base
  def to_s
    "  [#{id}] #{first} #{last}, <#{email}>, #{birthdate.strftime('%Y-%m-%d')}"
  end
...
```

Podemos destacar los siguientes puntos:
- se esta importando `sqlite3` que permite a Ruby interactuar con la base de datos SQLite
- tambien se esta importando la gema `active_record`que permite tratar bases de datos como clases de ruby
- En la siguiente linea hacemos la coneccion a la base de datos usando sqlite3, el archivo de la base de datos custumers.sqlite3
- se confirgura el ActiveRecord para que muestre las consultas SQL que se ejecutan en la consola
- se crea una clase llamada `Customer` que hereda de `ActiveRecord::Base`. Esto significa que `Customer` puede usar todas las caracteristicas de `ActiveRecord`, y cada instancia de `Customer` representa un registro de la tabla de clientes en la base de datos.
- se define el metodo `to_s` que se llama cuando se intenta convertir la instancia en una cadena.
